```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import LabelEncoder , MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense , LSTM , Dropout , Input
import re
import pickle
import tensorflow as tf 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import (Dense, Dropout, Input, Embedding, Bidirectional, GRU, Attention, GlobalAveragePooling1D)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping , ModelCheckpoint
from sklearn.metrics import classification_report
```

```python
fake = pd.read_csv('Fake.csv')
true = pd.read_csv('True.csv')
```

```python
fake['label']=1
true['label']=0
```

```python
true
```

```python
fake
```

```python
df = pd.concat([fake,true],axis=0)
```

```python
df.sample(frac=1,random_state=42)
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', "", text)          # remove URLs
    text = re.sub(r'[^a-zA-Z\s]', "", text)      # keep only letters + spaces
    text = re.sub(r"\s+", " ", text)             # collapse multiple spaces
    return text.strip()
```

```python
'                Vishwas                 '.strip()
```

```python
df['content']=df['text']+ " " + df['title']
```

```python
df['content'] = df['content'].apply(lambda x: clean_text(x))
```

```python
#Tokenization
```

```python
max_words = 50000
max_len = 500
```

```python
tokenizer = Tokenizer(num_words=max_words , oov_token='<OOV>')
```

```python
tokenizer.fit_on_texts(df['content'])
```

```python
len(tokenizer.word_index)
```

```python
X = tokenizer.texts_to_sequences(df['content'])
```

```python
max([max(i) for i in X if len(i)>1])
```

```python
X = pad_sequences(
    X,
    maxlen = max_len ,
    padding= 'post' ,
    truncating= 'post')
```

```python
X
```

```python
print(df.columns)
```

```python
y = df['label']
```

```python
y
```

```python
X
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

# Model

```python


inputs = Input(
    shape=(max_len,)
)

embedding = Embedding(
    input_dim= max_words,
    output_dim= 128
)(inputs)

gru1 = Bidirectional(
    GRU(128 ,return_sequences=True))(embedding)

gru2 = Bidirectional(
    GRU(64 ,return_sequences=True))(gru1)

attention = Attention()(
    [gru2,gru2]
)

pool = GlobalAveragePooling1D()(attention)

dense = Dense(64,activation='relu')(pool)

drop = Dropout(0.3)(dense)

outputs = Dense(1,activation='sigmoid')(drop)

model = Model(inputs , outputs) 
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
checkpoint = ModelCheckpoint('models/bigru_attention_model.h5',
                            save_best_only=True,
                            monitor='val_accuracy',
                             mode='max')
```

```python
earlystopping = EarlyStopping(patience=3 , restore_best_weights=True)
```

```python
history = model.fit(x_train , y_train , validation_split=0.2, epochs=5, batch_size=64,callbacks=[checkpoint , earlystopping])
```

```python
# EVALUATE
# ===================================

preds = (
    model.predict(x_test) > 0.5
).astype(int)

print(
    classification_report(
        y_test,
        preds
    )
)

# ===================================
# SAVE FINAL MODEL
# ===================================

model.save(
    "models/final_bigru_attention.keras"
)

# ===================================
# SAVE TOKENIZER
# ===================================

with open(
    "models/tokenizer.pkl",
    "wb"
) as f:

    pickle.dump(
        tokenizer,
        f
    )

print("\nTraining Complete")
print("Saved:")
print("models/bigru_attention_model.keras")
print("models/final_bigru_attention.keras")
print("models/tokenizer.pkl")
```
